```python
import numpy as np
import pandas as pd
```

```python
data1 = pd.read_csv("tmdb_5000_credits.csv")
data2 = pd.read_csv("tmdb_5000_movies.csv")
```

```python
data2c = data2.rename(columns ={"id":"movie_id"})
```

```python
data2c.head()
```

```python
data = pd.merge(data1,data2c,on='movie_id')
```

```python
data.head()
```

```python
data.columns
```

```python
from ast import literal_eval

featurenames = ['cast', 'crew', 'keywords', 'genres']
for feat in featurenames:
    data[feat] = data[feat].apply(literal_eval)
```

```python
def getdirector(b):
    for a in b:
        if a['job'] == 'Director':
            return a['name']
    return np.nan
```

```python
def getlist(b):
    if isinstance(b, list):
        names = [a['name'] for a in b]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
data['director'] = data['crew'].apply(getdirector)

featcols = ['cast', 'keywords', 'genres']
for feat in featcols:
    data[feat] = data[feat].apply(getlist)
```

```python
data[['title_x', 'cast', 'director', 'keywords', 'genres','overview']].head()
```

```python
def cleandata(b):
    if isinstance(b, list):
        return [str.lower(a.replace(" ", "")) for a in b]
    else:
        if isinstance(b, str):
            return str.lower(b.replace(" ", ""))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres','overview']

for feature in features:
    data[feature] = data[feature].apply(cleandata)
```

```python
def bagwords(b):
    return ' '.join(b['keywords']) + ' ' + ' '.join(b['cast']) + ' ' + b['director'] + ' ' + ' '.join(b['genres'])
data['bagwords'] = data.apply(bagwords, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
cvm = cv.fit_transform(data['bagwords'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(cvm, cvm)
```

```python
data = data.reset_index()
indices = pd.Series(data.index, index=data['title_x'])
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:16]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return data['title_x'].iloc[movie_indices]
```

```python
get_recommendations('Deadpool')
```
